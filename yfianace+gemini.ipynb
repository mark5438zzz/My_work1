```python
#最基礎版本的下載資料模塊
import yfinance as yf
# 股價走勢查詢函數
def get_stock_price_history(symbol, period="6mo"):
    try:
        stock = yf.Ticker(symbol)
        history = stock.history(period=period)
        history.reset_index(inplace=True)  # 確保日期作為一列而非索引
        history_records = history.tail().to_dict('records')
        return history_records
    except Exception as e:
        return f"無法獲取 {symbol} 的歷史價格: {e}"


# 股票資訊查詢函數
def get_stock_info(symbol):
    try:
        stock = yf.Ticker(symbol)
        info = stock.info
        return {
            '公司名稱': info.get('longName', 'N/A'),
            '現價': info.get('currentPrice', 'N/A'),
            '市值': info.get('marketCap', 'N/A'),
            '產業': info.get('industry', 'N/A'),
            '52週高點': info.get('fiftyTwoWeekHigh', 'N/A'),
            '52週低點': info.get('fiftyTwoWeekLow', 'N/A')
        }
    except:
        return f"無法獲取 {symbol} 的資訊"

# 基本財務數據查詢函數
def get_financial_data(symbol):
    try:
        stock = yf.Ticker(symbol)
        # 獲取最近的財務報表
        financials = stock.financials.iloc[:, 0]
        balance_sheet = stock.balance_sheet.iloc[:, 0]
        return {
            '營收': financials.get('Total Revenue', 'N/A'),
            '淨利': financials.get('Net Income', 'N/A'),
            '總資產': balance_sheet.get('Total Assets', 'N/A'),
            '總負債': balance_sheet.get('Total Liabilities Net Minority Interest', 'N/A'),
            'ROE': stock.info.get('returnOnEquity', 'N/A'),
            '本益比': stock.info.get('trailingPE', 'N/A')
        }
    except:
        return f"無法獲取 {symbol} 的財務數據"
```

```python
import yfinance as yf
import pandas as pd

def get_quarterly_financial_data(symbol):
    try:
        stock = yf.Ticker(symbol)
        # 獲取最近四個季度的財務報表
        financials = stock.financials
        balance_sheet = stock.balance_sheet

        quarterly_data = []
        for i in range(4):
            total_revenue = financials.loc['Total Revenue', :].iloc[i] if 'Total Revenue' in financials.index else None
            gross_profit = financials.loc['Gross Profit', :].iloc[i] if 'Gross Profit' in financials.index else None
            net_income = financials.loc['Net Income', :].iloc[i] if 'Net Income' in financials.index else None
            total_assets = balance_sheet.loc['Total Assets', :].iloc[i] if 'Total Assets' in balance_sheet.index else None
            total_liabilities = balance_sheet.loc['Total Liabilities Net Minority Interest', :].iloc[i] if 'Total Liabilities Net Minority Interest' in balance_sheet.index else None

            if total_revenue and isinstance(total_revenue, (int, float)) and total_revenue != 0 and gross_profit and isinstance(gross_profit, (int, float)):
                 gross_margin = (gross_profit / total_revenue) * 100
            else:
                 gross_margin = 'N/A'

            if total_assets and isinstance(total_assets, (int, float)) and total_liabilities and isinstance(total_liabilities, (int, float)):
                shareholder_equity = total_assets - total_liabilities
            else:
                shareholder_equity = None

            if shareholder_equity and isinstance(shareholder_equity, (int, float)) and shareholder_equity != 0 and net_income and isinstance(net_income, (int,float)):
                roe = (net_income / shareholder_equity) * 100
            else:
                roe = 'N/A'

            quarter_data = {
                '營收': total_revenue if total_revenue else 'N/A',
                '淨利': net_income if net_income else 'N/A',
                '總資產': total_assets if total_assets else 'N/A',
                '總負債': total_liabilities if total_liabilities else 'N/A',
                'ROE': roe,
                '本益比': stock.info.get('trailingPE', 'N/A'),
                '毛利率': gross_margin
            }
            quarterly_data.append(quarter_data)

        return quarterly_data
    except Exception as e:
        return f"無法獲取 {symbol} 的財務數據: {e}"
```

```python
#測試一整季的抓取(四份財報)
import pandas as pd
# 使用範例
symbol = "2355.TW"
financial_data = get_quarterly_financial_data(symbol)

if isinstance(financial_data, list):
    for i, data in enumerate(financial_data):
        print(f"第 {i+1} 季度數據:")
        for key, value in data.items():
            print(f"  {key}: {value}")
        print("-" * 20)
    df = pd.DataFrame(financial_data)
    print(df)
else:
    print(financial_data)
```

```python
def prepare_data(symbol):
    stock_info = get_stock_info(symbol)
    price_history = get_stock_price_history(symbol)
    financial_data = get_quarterly_financial_data(symbol)

    stock_info_str = f"""
    公司名稱: {stock_info.get('公司名稱')}
    現價: {stock_info.get('現價')}
    市值: {stock_info.get('市值')}
    產業: {stock_info.get('產業')}
    52週高點: {stock_info.get('52週高點')}
    52週低點: {stock_info.get('52週低點')}
    """

    # 確保日期鍵存在
    price_history_str = "\n".join(
        [f"日期: {record['Date']}, 收盤價: {record['Close']}" for record in price_history]
    ) if isinstance(price_history, list) else price_history

    financial_data_str = ""
    if isinstance(financial_data, list):
        for i, quarter_data in enumerate(financial_data):
             financial_data_str += f"""
             第 {i+1} 季度財務數據:
             營收: {quarter_data.get('營收', 'N/A')}
             淨利: {quarter_data.get('淨利', 'N/A')}
             總資產: {quarter_data.get('總資產', 'N/A')}
             總負債: {quarter_data.get('總負債', 'N/A')}
             ROE: {quarter_data.get('ROE', 'N/A')}
             本益比: {quarter_data.get('本益比', 'N/A')}
             毛利率: {quarter_data.get('毛利率', 'N/A')}
             """
    else:
         financial_data_str = f"無法獲取財務數據： {financial_data}"

    return f"股票資訊:\n{stock_info_str}\n價格歷史:\n{price_history_str}\n財務數據:\n{financial_data_str}"
```
